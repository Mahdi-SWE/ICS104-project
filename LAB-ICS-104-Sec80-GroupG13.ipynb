```python
def show(file):
    cardnumber=str(file)
    file=open(cardnumber+".txt","r")                      #user's file will be opened after he choose show account information
    line=file.readline()
    while line!='':
        line=line.strip()                                 #in this loop every line in the file will be read and printed to let 
        print(line)                                       #the user see the details
        line=file.readline()
        
        
    file.close()
    time.sleep(3)
```

```python
def changePINFun(currentPIN, cardNumber, file):
    File=open(str(file)+'.txt','r')                      #user's file will be opeded to change his PIN
    line=File.readline() 
    List=[]                                              #this list is to save user's information
    while line !='':
        line=line.strip()
        List.append(line)                                #while loop will read and save the information in the list
        line=File.readline()
    
    x=True
    while x:                                             #try and while function will check if the new PIN is acceptable or not
        try:
            new_PIN=int(input("Enter your new PIN "))
            while not len(str(new_PIN))==4: 
                    print("wrong input, you should enter 4 digits try again")   
                    new_PIN=int(input())
            x=False
        except ValueError:
            print("wrong input, you should only enter digits try again")
            
    List.pop(1)                                          #after the user enter correct new PIN the function will delete
    List.insert(1,new_PIN)                               #the old one and change it with the new one in the list that 
    new_file=open(str(cardNumber)+'.txt','w')            #was created earlier
    for i in List:
        new_file.write(str(i)+'\n')                      #this loop will write the information in the list into the user's 
                                                         #file because open function with mood write will delete everything
    new_file.close()
    time.sleep(3)
```

```python
def withdrawFun(money, cardNumber, file):
    
    File=open(str(file)+'.txt','r')                      #user's file will be opeded to deduct the amount that
    line=File.readline()                                 #he will withdraw it
    List=[]                                             
    while line !='':
        line=line.strip()                                #same in previous the list and while loop will save the information
        List.append(line)                                
        line=File.readline()
        
    balance=float(List[3])                               #the old balance will be deleted
    List.pop(3)
    
    while money>balance:                                       #this loop will start running if the user enter 
        print("you don't have enough money in your acount")    #bigger amount from what he has
        money=float(input("enter another amount "))
    
    balance=balance-money
    List.insert(3,str(balance))                                #after the balance updated it will be add to the information list
    x=time.asctime()
    List.append("withdraw at "+x+" by amount "+str(money))     #the detalise will be added to the list include the kind of
                                                               #the operation and the date with the time
    new_file=open(str(cardNumber)+'.txt','w')            
    for i in List:
        new_file.write(str(i)+'\n')                            #the information in the list will be saved in a new file with 
    new_file.close()                                           #the same card number of the user  
    
    time.sleep(3)
        
```

```python
def depositFun(file, nMoney, cardNumber):
    
    File=open(str(file)+'.txt','r')                 
    line=File.readline() 
    List=[]                                                  #the information will be stored in a list 
    while line !='':
        line=line.strip()
        List.append(line)                                
        line=File.readline()
        
    balance=float(List[3])
    List.pop(3)                                                  #the old balance will be deleted in the list. the balance will
    balance=balance+nMoney                                       #be updated after add the deposit and replace it with the 
    List.insert(3,str(balance))                                  #deleted then update the txt file with new list and inclode 
    x=time.asctime()                                             #the operation with the time and date in the txt
    List.append("deposit at "+x+" by amount "+str(nMoney))
    new_file=open(str(cardNumber)+'.txt','w')            
    for i in List:
        new_file.write(str(i)+'\n')                               
    new_file.close()
    
    time.sleep(3)
```

```python
def payBillFun(file, nMoney, cardNumber):  
    File=open(str(file)+'.txt','r')                 
    line=File.readline() 
    List=[]                                                      #the information will be stored in a list 
    while line !='':
        line=line.strip()
        List.append(line)                                
        line=File.readline()
    balance=float(List[3])   
    bill_name=input("Enter the name of the bill ")              
    bill_acount=input("Enter the account number of this bill ")  #the function will take the information for the bill
    bill_value=float(input("Enter the value of this bill "))
    
    if balance>=bill_value:                                      #if the account has enough balance, the balance will be
        List.pop(3)                                              #updated and replace in the lsit
        balance=balance-bill_value
        x=time.asctime()                                         #the detalise will be added to the list include the kind of
                                                                 #the operation and the date with the time
        List.insert(3,str(balance))
        List.append("pay bill with name "+bill_name+" and its number is "+str(bill_acount)+" at "+x+" by amount "+str(bill_value))
        
    else:
        print("there is no enough money in your account, please deposit amount to deduct the bill.")
        
    
    new_file=open(str(cardNumber)+'.txt','w')            
    for i in List:                                          #the updated information will be stored in txt with same number card
        new_file.write(str(i)+'\n')                      
    new_file.close()
    
    time.sleep(3)
```

```python
def viewTransactionsFun(cardNumber):
    File=open(str(cardNumber)+'.txt','r')                 
    line=File.readline() 
    List=[]                                                     #the information will be stored in a list 
    while line !='':
        line=line.strip()
        List.append(line)                                
        line=File.readline()
        
    if 'the Online payment is active' in List and len(List)>5:  #this if will check if the user active payment online to avoid 
        hestory=List[5:-1]                                      #print it with the transaction
        if len(hestory)>0:
            for i in hestory:                                   #this if will check if the user did more then one transactions
                print(i)                                            #they wll be printed
                                         
        if len(hestory)==0:
            print(List[5])
    elif 'the Online payment is active' not in List and len(List)>4:
        hestory=List[4:-1]                                      
        if len(hestory)>0:                                      #same here, function will check if the user did more tan one 
            for i in hestory:                                   #transaction or only one and show transaction\s to the user
                print(i)
        if len(hestory)==0:
            print(List[4])
    
    else:                                                     #if the user did not do transaction the function will print this
        print("no transactions")
        
    File.close()
    time.sleep(3)
```
